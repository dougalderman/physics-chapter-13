```r
x<-c(2.31, 2.01, 1.71, 1.34)
y<-c(2.55, 2.35, 2.15, 1.90)
```

```r
plot(x,y,col="blue", main="log(TE/T) vs log(RE/a)",xlab="log(TE/T) log(Earth's Orbital Period/Jupiter Moon Orbital Period", ylab="log(RE/a) log(Earth's Orbital Radius/Jupiter Moon Orbital Radius)")
```

```r
relation<-lm(y~I(x))
print(relation)
summary(relation)
```

```r
x_seq <- seq(from=1,to=3, by=0.01)
new_relation<-list(0)
for (i in 1:length(x_seq)) {
    x_new <- data.frame(x=x_seq[i])
    new_relation[i]<-predict(relation, x_new)
}
plot(x,y,col="blue", main="Linear Regression Example",lines(x_seq, new_relation),xlab="x",ylab="y")
```

```r
RSE_Percent = 0.001205/mean(y) * 100
```
